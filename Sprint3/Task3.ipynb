```
using System;
using System.Collections.Generic;


    // üîπ –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å
    public class SavingsAccount
    {
        // –ê—Ç—Ä–∏–±—É—Ç—ã
        public int AccountId { get; set; }
        public string OwnerName { get; set; }
        public decimal Balance { get; set; }
        public double InterestRate { get; set; }
        public string Currency { get; set; }
        public DateTime CreatedDate { get; set; }

        public SavingsAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency)
        {
            AccountId = accountId;
            OwnerName = ownerName;
            Balance = initialBalance;
            InterestRate = interestRate;
            Currency = currency;
            CreatedDate = DateTime.Now;
        }

        // –ú–µ—Ç–æ–¥—ã
        public virtual void Deposit(decimal amount)
        {
            Balance += amount;
            Console.WriteLine($"{OwnerName} –≤–Ω–µ—Å {amount} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance} {Currency}");
        }

        // üîπ –ü–µ—Ä–µ–≥—Ä—É–∑–∫–∞ –º–µ—Ç–æ–¥–∞ Deposit
        public void Deposit(decimal amount, string comment)
        {
            Deposit(amount);
            Console.WriteLine($"–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π: {comment}");
        }

        public virtual void Withdraw(decimal amount)
        {
            if (amount <= Balance)
            {
                Balance -= amount;
                Console.WriteLine($"{OwnerName} —Å–Ω—è–ª {amount} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance}");
            }
            else
            {
                Console.WriteLine("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å—Ä–µ–¥—Å—Ç–≤!");
            }
        }

        public virtual void CalculateInterest()
        {
            decimal interest = Balance * (decimal)(InterestRate / 100);
            Balance += interest;
            Console.WriteLine($"–ù–∞—á–∏—Å–ª–µ–Ω—ã –ø—Ä–æ—Ü–µ–Ω—Ç—ã {interest} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance}");
        }

        public void ShowInfo()
        {
            Console.WriteLine($"[{AccountId}] {OwnerName} | –ë–∞–ª–∞–Ω—Å: {Balance} {Currency} | –°—Ç–∞–≤–∫–∞: {InterestRate}% | –°–æ–∑–¥–∞–Ω: {CreatedDate}");
        }
    }

    // üîπ 1. –°—Ç—É–¥–µ–Ω—á–µ—Å–∫–∏–π —Å—á–µ—Ç
    public class StudentAccount : SavingsAccount
    {
        public int YearOfStudy { get; set; }
        public string University { get; set; }
        public string StudentId { get; set; }
        public bool ScholarshipActive { get; set; }

        public StudentAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency,
                              int yearOfStudy, string university, string studentId)
            : base(accountId, ownerName, initialBalance, interestRate, currency)
        {
            YearOfStudy = yearOfStudy;
            University = university;
            StudentId = studentId;
            ScholarshipActive = true;
        }

        // üîπ –ü–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ –º–µ—Ç–æ–¥–∞
        public override void CalculateInterest()
        {
            double reducedRate = InterestRate / 2;
            decimal interest = Balance * (decimal)(reducedRate / 100);
            Balance += interest;
            Console.WriteLine($"–°—Ç—É–¥–µ–Ω—Ç—É {OwnerName} –Ω–∞—á–∏—Å–ª–µ–Ω—ã {interest} {Currency} (—Å—Ç–∞–≤–∫–∞ {reducedRate}%). –ë–∞–ª–∞–Ω—Å: {Balance}");
        }

        // –ù–æ–≤—ã–µ –º–µ—Ç–æ–¥—ã
        public void GetScholarship(decimal amount)
        {
            if (ScholarshipActive)
            {
                Deposit(amount);
                Console.WriteLine($"–°—Ç–∏–ø–µ–Ω–¥–∏—è {amount} {Currency} –∑–∞—á–∏—Å–ª–µ–Ω–∞.");
            }
        }

        public void FreezeScholarship()
        {
            ScholarshipActive = false;
            Console.WriteLine("–°—Ç–∏–ø–µ–Ω–¥–∏—è –ø—Ä–∏–æ—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞.");
        }

        public void ActivateScholarship()
        {
            ScholarshipActive = true;
            Console.WriteLine("–°—Ç–∏–ø–µ–Ω–¥–∏—è –∞–∫—Ç–∏–≤–∏—Ä–æ–≤–∞–Ω–∞.");
        }
    }

    // üîπ 2. –ü—Ä–µ–º–∏—É–º —Å—á–µ—Ç
    public class PremiumAccount : SavingsAccount
    {
        public decimal MinimumBalance { get; set; }
        public int FreeTransfers { get; set; }
        public string StatusLevel { get; set; }
        public List<string> Privileges { get; set; }

        public PremiumAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency,
                              decimal minimumBalance)
            : base(accountId, ownerName, initialBalance, interestRate, currency)
        {
            MinimumBalance = minimumBalance;
            FreeTransfers = 5;
            StatusLevel = "Gold";
            Privileges = new List<string> { "–ö–æ–Ω—Å—å–µ—Ä–∂-—Å–µ—Ä–≤–∏—Å", "–ö—ç—à–±—ç–∫" };
        }

        // üîπ –ü–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ –º–µ—Ç–æ–¥–∞
        public override void Withdraw(decimal amount)
        {
            if (Balance - amount >= MinimumBalance)
            {
                Balance -= amount;
                Console.WriteLine($"{OwnerName} (Premium) —Å–Ω—è–ª {amount} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance}");
            }
            else
            {
                Console.WriteLine($"–ù–µ–ª—å–∑—è —Å–Ω—è—Ç—å {amount}, –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å –¥–æ–ª–∂–µ–Ω –æ—Å—Ç–∞–≤–∞—Ç—å—Å—è {MinimumBalance}");
            }
        }

        // –ù–æ–≤—ã–µ –º–µ—Ç–æ–¥—ã
        public void ShowPrivileges()
        {
            Console.WriteLine($"–ü—Ä–∏–≤–∏–ª–µ–≥–∏–∏ {OwnerName}: {string.Join(", ", Privileges)}");
        }

        public void UpgradeStatus(string newLevel)
        {
            StatusLevel = newLevel;
            Console.WriteLine($"{OwnerName} –ø–æ–ª—É—á–∏–ª –Ω–æ–≤—ã–π —Å—Ç–∞—Ç—É—Å: {StatusLevel}");
        }

        public void UseFreeTransfer()
        {
            if (FreeTransfers > 0)
            {
                FreeTransfers--;
                Console.WriteLine($"–ë–µ—Å–ø–ª–∞—Ç–Ω—ã–π –ø–µ—Ä–µ–≤–æ–¥ –≤—ã–ø–æ–ª–Ω–µ–Ω. –û—Å—Ç–∞–ª–æ—Å—å: {FreeTransfers}");
            }
            else
            {
                Console.WriteLine("–ë–µ—Å–ø–ª–∞—Ç–Ω—ã–µ –ø–µ—Ä–µ–≤–æ–¥—ã –∑–∞–∫–æ–Ω—á–∏–ª–∏—Å—å!");
            }
        }
    }

    // üîπ 3. –ò–Ω–≤–µ—Å—Ç–∏—Ü–∏–æ–Ω–Ω—ã–π —Å—á–µ—Ç
    public class InvestmentAccount : SavingsAccount
    {
        public decimal PortfolioValue { get; private set; }
        public int RiskLevel { get; set; }
        public string Strategy { get; set; }
        public bool AutoInvestEnabled { get; set; }

        public InvestmentAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency,
                                 decimal portfolioValue, int riskLevel, string strategy)
            : base(accountId, ownerName, initialBalance, interestRate, currency)
        {
            PortfolioValue = portfolioValue;
            RiskLevel = riskLevel;
            Strategy = strategy;
            AutoInvestEnabled = true;
        }

        // üîπ –ü–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ –º–µ—Ç–æ–¥–∞
        public override void Deposit(decimal amount)
        {
            decimal investPart = AutoInvestEnabled ? amount * 0.3m : 0;
            decimal depositPart = amount - investPart;
            Balance += depositPart;
            PortfolioValue += investPart;
            Console.WriteLine($"–ò–Ω–≤–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–æ {investPart}, –∑–∞—á–∏—Å–ª–µ–Ω–æ {depositPart}. –ë–∞–ª–∞–Ω—Å: {Balance}, –ü–æ—Ä—Ç—Ñ–µ–ª—å: {PortfolioValue}");
        }

        // –ù–æ–≤—ã–µ –º–µ—Ç–æ–¥—ã
        public void RebalancePortfolio()
        {
            Console.WriteLine($"–ü–æ—Ä—Ç—Ñ–µ–ª—å {OwnerName} —Ä–µ–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω –ø–æ —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏: {Strategy}");
        }

        public void ChangeRisk(int newRisk)
        {
            RiskLevel = newRisk;
            Console.WriteLine($"–†–∏—Å–∫ –∏–∑–º–µ–Ω–µ–Ω –Ω–∞ {RiskLevel}");
        }

        public void ToggleAutoInvest()
        {
            AutoInvestEnabled = !AutoInvestEnabled;
            Console.WriteLine($"–ê–≤—Ç–æ–∏–Ω–≤–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ: {(AutoInvestEnabled ? "–≤–∫–ª—é—á–µ–Ω–æ" : "–≤—ã–∫–ª—é—á–µ–Ω–æ")}");
        }
    }

    // üîπ Generic –∫–ª–∞—Å—Å
    public class AccountManager<T> where T : SavingsAccount
    {
        private List<T> accounts = new List<T>();

        public void AddAccount(T account) => accounts.Add(account);

        public void ShowAllAccounts()
        {
            foreach (var acc in accounts)
            {
                acc.ShowInfo();
            }
        }

        public T FindById(int id)
        {
            return accounts.Find(a => a.AccountId == id);
        }

        public void Transfer(T from, T to, decimal amount)
        {
            from.Withdraw(amount);
            to.Deposit(amount, "–ü–µ—Ä–µ–≤–æ–¥ —Å—Ä–µ–¥—Å—Ç–≤");
        }
    }

            StudentAccount student = new StudentAccount(1, "–ò–≤–∞–Ω", 1000m, 5, "RUB", 2, "–¢—é–º–ì–£", "ST123");
            PremiumAccount premium = new PremiumAccount(2, "–ê–Ω–Ω–∞", 5000m, 7, "RUB", 1000m);
            InvestmentAccount invest = new InvestmentAccount(3, "–û–ª–µ–≥", 15000m, 8, "USD", 3000m, 2, "–ê–≥—Ä–µ—Å—Å–∏–≤–Ω–∞—è");

            // –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è –º–µ—Ç–æ–¥–æ–≤
            student.GetScholarship(2000);
            student.CalculateInterest();

            premium.Withdraw(4500);
            premium.ShowPrivileges();
            premium.UpgradeStatus("Platinum");

            invest.Deposit(5000);
            invest.RebalancePortfolio();
            invest.ChangeRisk(3);
            invest.ToggleAutoInvest();

            // –†–∞–±–æ—Ç–∞ generic-–∫–ª–∞—Å—Å–∞
            AccountManager<SavingsAccount> manager = new AccountManager<SavingsAccount>();
            manager.AddAccount(student);
            manager.AddAccount(premium);
            manager.AddAccount(invest);

            Console.WriteLine("\n--- –í—Å–µ —Å—á–µ—Ç–∞ ---");
            manager.ShowAllAccounts();

            Console.WriteLine("\n--- –ü–µ—Ä–µ–≤–æ–¥ —á–µ—Ä–µ–∑ generic-–∫–ª–∞—Å—Å ---");
            manager.Transfer(premium, student, 1000);
```
