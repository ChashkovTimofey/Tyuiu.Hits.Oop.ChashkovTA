```
using System;
using System.Collections.Generic;


    // üîπ –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å
    public class SavingsAccount
    {
        public int AccountId { get; set; }
        public string OwnerName { get; set; }
        public decimal Balance { get; set; }
        public double InterestRate { get; set; }
        public DateTime CreatedDate { get; set; }
        public string Currency { get; set; }

        public SavingsAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency)
        {
            AccountId = accountId;
            OwnerName = ownerName;
            Balance = initialBalance;
            InterestRate = interestRate;
            CreatedDate = DateTime.Now;
            Currency = currency;
        }

        public virtual void Deposit(decimal amount)
        {
            Balance += amount;
            Console.WriteLine($"{OwnerName} –≤–Ω–µ—Å {amount} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance} {Currency}");
        }

        public virtual void Withdraw(decimal amount)
        {
            if (amount <= Balance)
            {
                Balance -= amount;
                Console.WriteLine($"{OwnerName} —Å–Ω—è–ª {amount} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance} {Currency}");
            }
            else
            {
                Console.WriteLine("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å—Ä–µ–¥—Å—Ç–≤!");
            }
        }

        public virtual void CalculateInterest()
        {
            decimal interest = Balance * (decimal)(InterestRate / 100);
            Balance += interest;
            Console.WriteLine($"{OwnerName} –Ω–∞—á–∏—Å–ª–µ–Ω—ã –ø—Ä–æ—Ü–µ–Ω—Ç—ã {interest} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance} {Currency}");
        }

        // –î–æ–ø. –º–µ—Ç–æ–¥—ã
        public void ShowInfo()
        {
            Console.WriteLine($"[{AccountId}] –í–ª–∞–¥–µ–ª–µ—Ü: {OwnerName}, –ë–∞–ª–∞–Ω—Å: {Balance} {Currency}, –°—Ç–∞–≤–∫–∞: {InterestRate}%");
        }

        public void ChangeCurrency(string newCurrency, decimal exchangeRate)
        {
            Balance *= exchangeRate;
            Currency = newCurrency;
            Console.WriteLine($"–ë–∞–ª–∞–Ω—Å –ø–µ—Ä–µ–≤–µ–¥–µ–Ω –≤ {Currency}: {Balance}");
        }

        public void TransferTo(SavingsAccount targetAccount, decimal amount)
        {
            if (amount <= Balance)
            {
                this.Withdraw(amount);
                targetAccount.Deposit(amount);
                Console.WriteLine($"–ü–µ—Ä–µ–≤–µ–¥–µ–Ω–æ {amount} {Currency} —Å–æ —Å—á–µ—Ç–∞ {AccountId} –Ω–∞ —Å—á–µ—Ç {targetAccount.AccountId}");
            }
            else
            {
                Console.WriteLine($"–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å—Ä–µ–¥—Å—Ç–≤ –¥–ª—è –ø–µ—Ä–µ–≤–æ–¥–∞ {amount}!");
            }
        }
    }

    // üîπ –ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å –¥–ª—è –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è
    public interface IRewardable
    {
        void AddRewardPoints(int points);
        void RedeemPoints(int points);
    }

    // üîπ 1. –°—Ç—É–¥–µ–Ω—á–µ—Å–∫–∏–π —Å—á–µ—Ç (–ø—Ä–æ—Å—Ç–æ–µ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ)
    public class StudentAccount : SavingsAccount
    {
        public int YearOfStudy { get; set; }
        public string University { get; set; }
        public string StudentId { get; set; }

        public StudentAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency,
                              int yearOfStudy, string university, string studentId)
            : base(accountId, ownerName, initialBalance, interestRate, currency)
        {
            YearOfStudy = yearOfStudy;
            University = university;
            StudentId = studentId;
        }

        public override void CalculateInterest()
        {
            double reducedRate = InterestRate / 2;
            decimal interest = Balance * (decimal)(reducedRate / 100);
            Balance += interest;
            Console.WriteLine($"–°—Ç—É–¥–µ–Ω—Ç {OwnerName} –ø–æ–ª—É—á–∏–ª –ø—Ä–æ—Ü–µ–Ω—Ç—ã {interest} {Currency} (—Å—Ç–∞–≤–∫–∞ {reducedRate}%). –ë–∞–ª–∞–Ω—Å: {Balance}");
        }

        public void GetScholarship(decimal amount)
        {
            Deposit(amount);
            Console.WriteLine($"–°—Ç–∏–ø–µ–Ω–¥–∏—è {amount} {Currency} –∑–∞—á–∏—Å–ª–µ–Ω–∞ —Å—Ç—É–¥–µ–Ω—Ç—É {OwnerName}");
        }

        public void PayTuition(decimal amount)
        {
            Withdraw(amount);
            Console.WriteLine($"{OwnerName} –æ–ø–ª–∞—Ç–∏–ª –æ–±—É—á–µ–Ω–∏–µ –Ω–∞ —Å—É–º–º—É {amount} {Currency}");
        }
    }

    // üîπ 2. –ü—Ä–µ–º–∏—É–º —Å—á–µ—Ç (—Å–ª–æ–∂–Ω–æ–µ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ ‚Üí –æ—Ç –Ω–µ–≥–æ —É–Ω–∞—Å–ª–µ–¥—É–µ–º –µ—â—ë –æ–¥–∏–Ω –∫–ª–∞—Å—Å)
    public class PremiumAccount : SavingsAccount
    {
        public decimal MinimumBalance { get; set; }
        public List<string> Privileges { get; set; }

        public PremiumAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency,
                              decimal minimumBalance)
            : base(accountId, ownerName, initialBalance, interestRate, currency)
        {
            MinimumBalance = minimumBalance;
            Privileges = new List<string> { "–ö–æ–Ω—Å—å–µ—Ä–∂-—Å–µ—Ä–≤–∏—Å", "–ë–µ—Å–ø–ª–∞—Ç–Ω—ã–µ –ø–µ—Ä–µ–≤–æ–¥—ã", "–ö—ç—à–±—ç–∫" };
        }

        public override void Withdraw(decimal amount)
        {
            if (Balance - amount >= MinimumBalance)
            {
                Balance -= amount;
                Console.WriteLine($"–° –ø—Ä–µ–º–∏—É–º-—Å—á–µ—Ç–∞ {OwnerName} —Å–Ω—è—Ç–æ {amount} {Currency}. –ë–∞–ª–∞–Ω—Å: {Balance}");
            }
            else
            {
                Console.WriteLine($"–ù–µ–ª—å–∑—è —Å–Ω—è—Ç—å {amount}, –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å –¥–æ–ª–∂–µ–Ω –æ—Å—Ç–∞–≤–∞—Ç—å—Å—è {MinimumBalance} {Currency}.");
            }
        }

        public void ShowPrivileges()
        {
            Console.WriteLine($"–ü—Ä–∏–≤–∏–ª–µ–≥–∏–∏ {OwnerName}: {string.Join(", ", Privileges)}");
        }

        public void AddPrivilege(string privilege)
        {
            Privileges.Add(privilege);
            Console.WriteLine($"–î–æ–±–∞–≤–ª–µ–Ω–∞ –Ω–æ–≤–∞—è –ø—Ä–∏–≤–∏–ª–µ–≥–∏—è: {privilege}");
        }
    }

    // üîπ 2.1. VIP-—Å—á–µ—Ç (–Ω–∞—Å–ª–µ–¥–Ω–∏–∫ PremiumAccount) ‚Üí —Å–ª–æ–∂–Ω–æ–µ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ
    public class VIPAccount : PremiumAccount, IRewardable
    {
        public int RewardPoints { get; private set; }
        public string PersonalManager { get; set; }

        public VIPAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency,
                          decimal minimumBalance, string manager)
            : base(accountId, ownerName, initialBalance, interestRate, currency, minimumBalance)
        {
            PersonalManager = manager;
            RewardPoints = 0;
        }

        public void AddRewardPoints(int points)
        {
            RewardPoints += points;
            Console.WriteLine($"{OwnerName} –ø–æ–ª—É—á–∏–ª {points} –±–æ–Ω—É—Å–Ω—ã—Ö –±–∞–ª–ª–æ–≤. –í—Å–µ–≥–æ: {RewardPoints}");
        }

        public void RedeemPoints(int points)
        {
            if (RewardPoints >= points)
            {
                RewardPoints -= points;
                Console.WriteLine($"{OwnerName} –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª {points} –±–∞–ª–ª–æ–≤. –û—Å—Ç–∞—Ç–æ–∫: {RewardPoints}");
            }
            else
            {
                Console.WriteLine("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–∞–ª–ª–æ–≤ –¥–ª—è —Å–ø–∏—Å–∞–Ω–∏—è!");
            }
        }

        public void ContactManager()
        {
            Console.WriteLine($"{OwnerName} —Å–≤—è–∑–∞–ª—Å—è —Å –ª–∏—á–Ω—ã–º –º–µ–Ω–µ–¥–∂–µ—Ä–æ–º: {PersonalManager}");
        }
    }

    // üîπ 3. –ò–Ω–≤–µ—Å—Ç–∏—Ü–∏–æ–Ω–Ω—ã–π —Å—á–µ—Ç (–º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ —á–µ—Ä–µ–∑ –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å)
    public class InvestmentAccount : SavingsAccount, IRewardable
    {
        public decimal PortfolioValue { get; private set; }
        public int RiskLevel { get; set; }
        public string InvestmentStrategy { get; set; }
        public int RewardPoints { get; private set; }

        public InvestmentAccount(int accountId, string ownerName, decimal initialBalance, double interestRate, string currency,
                                 decimal portfolioValue, int riskLevel, string strategy)
            : base(accountId, ownerName, initialBalance, interestRate, currency)
        {
            PortfolioValue = portfolioValue;
            RiskLevel = riskLevel;
            InvestmentStrategy = strategy;
            RewardPoints = 0;
        }

        public override void Deposit(decimal amount)
        {
            decimal investPart = amount * 0.3m;
            decimal depositPart = amount - investPart;
            Balance += depositPart;
            PortfolioValue += investPart;
            Console.WriteLine($"–ò–Ω–≤–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–æ {investPart}, –Ω–∞ —Å—á–µ—Ç –∑–∞—á–∏—Å–ª–µ–Ω–æ {depositPart}. –ë–∞–ª–∞–Ω—Å: {Balance}, –ü–æ—Ä—Ç—Ñ–µ–ª—å: {PortfolioValue}");
        }

        public void RebalancePortfolio()
        {
            Console.WriteLine($"{OwnerName} –ø—Ä–æ–≤–µ–ª —Ä–µ–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫—É –ø–æ—Ä—Ç—Ñ–µ–ª—è –ø–æ —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ {InvestmentStrategy}");
        }

        public void ChangeRiskLevel(int newRisk)
        {
            RiskLevel = newRisk;
            Console.WriteLine($"–£—Ä–æ–≤–µ–Ω—å —Ä–∏—Å–∫–∞ –∏–∑–º–µ–Ω–µ–Ω –Ω–∞ {RiskLevel}");
        }

        public void AddRewardPoints(int points)
        {
            RewardPoints += points;
            Console.WriteLine($"{OwnerName} –ø–æ–ª—É—á–∏–ª {points} –±–æ–Ω—É—Å–æ–≤. –í—Å–µ–≥–æ: {RewardPoints}");
        }

        public void RedeemPoints(int points)
        {
            if (RewardPoints >= points)
            {
                RewardPoints -= points;
                Console.WriteLine($"{OwnerName} –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª {points} –±–æ–Ω—É—Å–æ–≤. –û—Å—Ç–∞—Ç–æ–∫: {RewardPoints}");
            }
            else
            {
                Console.WriteLine("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–Ω—É—Å–æ–≤!");
            }
        }
    }


            StudentAccount student = new StudentAccount(1, "–ò–≤–∞–Ω", 1000m, 5, "RUB", 2, "–¢—é–º–ì–£", "ST123");
            PremiumAccount premium = new PremiumAccount(2, "–ê–Ω–Ω–∞", 5000m, 7, "RUB", 1000m);
            VIPAccount vip = new VIPAccount(3, "–û–ª–µ–≥", 20000m, 10, "USD", 5000m, "–ú–∞—Ä–∏—è –ü–µ—Ç—Ä–æ–≤–∞");
            InvestmentAccount invest = new InvestmentAccount(4, "–°–µ—Ä–≥–µ–π", 15000m, 8, "USD", 3000m, 2, "–ê–≥—Ä–µ—Å—Å–∏–≤–Ω–∞—è");

            Console.WriteLine("\n--- –°—Ç—É–¥–µ–Ω—á–µ—Å–∫–∏–π —Å—á–µ—Ç ---");
            student.ShowInfo();
            student.GetScholarship(3000);
            student.PayTuition(1500);
            student.CalculateInterest();

            Console.WriteLine("\n--- –ü—Ä–µ–º–∏—É–º —Å—á–µ—Ç ---");
            premium.ShowInfo();
            premium.Withdraw(4500);
            premium.ShowPrivileges();

            Console.WriteLine("\n--- VIP —Å—á–µ—Ç ---");
            vip.ShowInfo();
            vip.ContactManager();
            vip.AddRewardPoints(500);
            vip.RedeemPoints(200);

            Console.WriteLine("\n--- –ò–Ω–≤–µ—Å—Ç–∏—Ü–∏–æ–Ω–Ω—ã–π —Å—á–µ—Ç ---");
            invest.ShowInfo();
            invest.Deposit(5000);
            invest.RebalancePortfolio();
            invest.ChangeRiskLevel(3);
            invest.AddRewardPoints(300);
```
